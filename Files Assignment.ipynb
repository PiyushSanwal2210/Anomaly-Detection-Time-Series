Question = 1 >>> What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples?

Ans = Anomaly Detection (also called outlier detection) is the process of identifying data points, events, or patterns that deviate significantly from the expected behavior in a dataset.

These anomalies often indicate:

Fraudulent activity (credit card fraud, insurance fraud)

Faults (manufacturing defects, sensor failures)

Intrusions (cybersecurity threats)

Rare diseases or unusual health conditions in medical data

üìå Types of Anomalies
1. Point Anomalies

A single data point is anomalous if it is too far from the rest of the data.

Most common type of anomaly.

Often detected using distance-based methods (e.g., z-scores, DBSCAN, isolation forests).

‚úÖ Example:

A customer suddenly spends ‚Çπ10,00,000 when their usual spending is around ‚Çπ5,000 per month.

A temperature sensor normally reads around 25¬∞C, but suddenly shows 100¬∞C.

2. Contextual Anomalies (also called conditional anomalies)

A data point is anomalous in a specific context, but may be normal in another.

Context is defined by time, location, or situation.

‚úÖ Example:

A temperature of 30¬∞C is normal in summer but anomalous in winter.

A person logging in from New York is normal, but logging in from New York and London within 5 minutes is suspicious.

Heart rate of 120 bpm is normal during exercise but abnormal when resting.

3. Collective Anomalies

A group of related data points together is anomalous, even if individual points look normal.

Usually detected in time-series or sequential data.

‚úÖ Example:

Sudden spike of network traffic for 10 minutes indicating a DDoS attack.

A stock price showing a sudden sustained downward trend (indicating a market crash).

ECG readings: one abnormal sequence of beats could indicate arrhythmia, even if each reading alone seems fine.

üìù Summary Table
Type	Definition	Example
Point anomaly	Single data point deviates from others	Credit card transaction of ‚Çπ10,00,000
Contextual anomaly	Anomalous in a specific context	30¬∞C temperature in winter
Collective anomaly	A group of points together is abnormal	DDoS attack traffic pattern

‚úÖ In real-world anomaly detection systems (fraud detection, intrusion detection, predictive maintenance), all three types may occur, and models must be tailored to handle them.

Question = 2 >>> Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases ?

Ans = 1. Isolation Forest (IF)
# Approach:

Based on the idea that anomalies are easier to isolate than normal points.

Builds random decision trees and measures the number of splits required to isolate a point.

Few splits = likely anomaly.

# Suitable Use Cases:

High-dimensional datasets (works well even with many features).

Large datasets (scales efficiently).

Generic anomaly detection without requiring assumptions about data shape.

# Limitations:

Assumes anomalies are sparse and few.

May miss anomalies if they are not well-separated.

# 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
# Approach:

A density-based clustering method.

Defines clusters as areas with high density and marks points in low-density areas as noise (anomalies).

Controlled by parameters eps (neighborhood radius) and min_samples.

# Suitable Use Cases:

Spatial data (geographic, location-based).

Detecting clusters of arbitrary shape (e.g., circles, blobs).

Identifying noise/outliers in 2D or 3D data.

# Limitations:

Struggles in high-dimensional data (curse of dimensionality).

Sensitive to parameter choice (eps, min_samples).

# 3. Local Outlier Factor (LOF)
# Approach:

Measures local density deviation of a data point relative to its neighbors.

If a point‚Äôs density is much lower than its neighbors‚Äô, it is an outlier.

Focuses on local context rather than global distribution.

# Suitable Use Cases:

Data where local anomalies exist (contextual anomalies).

Situations where clusters have varying densities.

Small-to-medium datasets with interpretable neighborhood structures.

# Limitations:

Computationally expensive for large datasets.

Not suitable for very high-dimensional data.

# Comparison Table
Method	Approach	Works Best On	Weaknesses
Isolation Forest	Random partitioning, anomalies isolate faster	High-dimensional, large datasets	May miss subtle anomalies
DBSCAN	Density-based clustering, noise = anomalies	Low-dimensional, spatial data, arbitrary cluster shapes	Poor in high dimensions, parameter sensitive
LOF	Local density comparison (neighbors)	Local/contextual anomalies, varying-density clusters	Slow on large datasets, not scalable

# Rule of Thumb:

Use Isolation Forest for high-dimensional and large datasets (fraud detection, log data).

Use DBSCAN for spatial or clustering + anomaly detection tasks (GPS data, image segmentation).

Use LOF when local context matters (detecting unusual user behavior in a subgroup).


Question = 3 >>> What are the key components of a Time Series? Explain each with one
example?

Ans = Key Components of a Time Series

A time series is a sequence of observations taken at regular time intervals (daily, monthly, yearly, etc.). It usually has four main components:

1. Trend (T)

# Definition:
The long-term upward or downward movement in the data over time.
It reflects the overall direction of the series, ignoring short-term fluctuations.

# Example:

The number of internet users worldwide has been steadily increasing over the past 20 years.

Stock market indices (like S&P 500) show a long-term upward trend despite short-term ups and downs.

2. Seasonality (S)

# Definition:
Regular, repeating patterns at fixed intervals (daily, weekly, monthly, yearly).
Caused by calendar or seasonal factors.

# Example:

Ice cream sales peak every summer and drop in winter.

Electricity demand is higher during daytime and lower at night.

3. Cyclic Component (C)

# Definition:
Long-term fluctuations that occur due to economic, business, or natural cycles.
Unlike seasonality, cycles do not have a fixed period and may last several years.

# Example:

Economic cycles: Periods of boom and recession affect stock prices and employment rates.

Real estate market: Housing prices may rise and fall every 8‚Äì10 years.

4. Irregular / Random Component (R)

# Definition:
Unpredictable, random variations caused by unusual or one-time events.
These cannot be explained by trend, seasonality, or cycles.

# Example:

A sudden earthquake disrupts power consumption data.

A pandemic (COVID-19) caused unexpected spikes in medical supply demand.

A company‚Äôs stock price crashes due to an unexpected scandal.

# Summary Table
Component	Definition	Example
Trend (T)	Long-term direction	Internet users increasing over years
Seasonality (S)	Regular, repeating patterns	Ice cream sales higher in summer
Cycle (C)	Long-term fluctuations (not fixed)	Economic boom and recession
Irregular (R)	Random, unpredictable variations	COVID-19 impact on travel data

Question = 4 >>> Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

Ans = A time series is stationary if its statistical properties do not change over time.

That means:

Mean is constant over time

Variance is constant (homoscedasticity)

Autocovariance depends only on the lag between observations, not on actual time

In other words, a stationary series looks stable and predictable in terms of its distribution.

# Example

Daily returns of a stock are usually stationary (fluctuate around a constant mean, with constant variance).

Stock prices themselves are typically non-stationary (they trend upward or downward over time).

# How to Test for Stationarity
1. Visual Inspection

Plot the time series.

If you see a trend or changing variance, it‚Äôs likely non-stationary.

2. Summary Statistics

Divide the series into parts and compare the mean and variance.

If they change significantly, the series is non-stationary.

3. Statistical Tests

Augmented Dickey-Fuller (ADF) Test

Null Hypothesis (H‚ÇÄ): The series has a unit root (non-stationary).

If p-value < 0.05 ‚Üí reject H‚ÇÄ ‚Üí series is stationary.

KPSS Test (Kwiatkowski‚ÄìPhillips‚ÄìSchmidt‚ÄìShin)

Null Hypothesis (H‚ÇÄ): The series is stationary.

If p-value < 0.05 ‚Üí reject H‚ÇÄ ‚Üí series is non-stationary.

# Often used together for confirmation.

# 3. ARIMA (Autoregressive Integrated Moving Average)
# Structure:

Combines AR + I (Integration = differencing for stationarity) + MA.

p = AR terms

q = MA terms

# Application:

Handles non-stationary series (trend + noise).

Example: Forecasting monthly sales where sales increase steadily (trend) but have short-term fluctuations.

# Structure:

Extends ARIMA to handle seasonality explicitly.

General form: SARIMA(p, d, q)(P, D, Q, s)

(p, d, q) = non-seasonal components

(P, D, Q) = seasonal AR, differencing, MA

s = seasonal period (e.g., 12 for monthly data with yearly seasonality)

# Application:

Suitable for seasonal data.

Example: Forecasting ice cream sales (peaks every summer) or electricity demand (daily/weekly cycles).

Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components
(Include your Python code and output in the code box below.)

```python
# ================================
# Time Series Decomposition Example
# Using AirPassengers Dataset
# ================================

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# -----------------------
# 1. Load the Dataset
# -----------------------
# Dataset: Monthly totals of international airline passengers, 1949 to 1960
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# -----------------------
# 2. Plot Original Series
# -----------------------
plt.figure(figsize=(10, 4))
plt.plot(data, label="AirPassengers")
plt.title("Monthly Air Passengers (1949-1960)")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# -----------------------
# 3. Seasonal Decomposition
# -----------------------
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot decomposition
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

Question = 7 >>> : Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

```python
# ==============================================
# Isolation Forest Anomaly Detection (Synthetic Data)
# ==============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# -----------------------
# 1. Generate Synthetic Data (mimicking fares vs distance)
# -----------------------
X, _ = make_blobs(n_samples=400, centers=[(5, 20)], cluster_std=[5], random_state=42)

# Create DataFrame with columns similar to taxi data
df = pd.DataFrame(X, columns=["trip_distance", "fare_amount"])

# Add some anomalies manually
rng = np.random.RandomState(42)
anomalies = rng.uniform(low=0, high=30, size=(20, 2))
df = pd.concat([df, pd.DataFrame(anomalies, columns=["trip_distance", "fare_amount"])])

# -----------------------
# 2. Apply Isolation Forest
# -----------------------
iso = IsolationForest(contamination=0.05, random_state=42)
df['anomaly'] = iso.fit_predict(df[['trip_distance', 'fare_amount']])

# -----------------------
# 3. Visualization
# -----------------------
plt.figure(figsize=(10,6))

# Normal points
plt.scatter(df['trip_distance'][df['anomaly'] == 1],
            df['fare_amount'][df['anomaly'] == 1],
            c='blue', label='Normal', alpha=0.6)

# Anomalies
plt.scatter(df['trip_distance'][df['anomaly'] == -1],
            df['fare_amount'][df['anomaly'] == -1],
            c='red', label='Anomaly', alpha=0.8, marker='x')

plt.title("Isolation Forest - Anomaly Detection (Synthetic Taxi Data)")
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.legend()
plt.show()
```

Question = 8 >>> Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)

```python
# ==============================================
# SARIMA Forecasting on AirPassengers Dataset
# ==============================================

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------
# 1. Load Dataset
# -----------------------
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
ts = data['Passengers']

# -----------------------
# 2. Fit SARIMA Model
# -----------------------
# (p,d,q) x (P,D,Q,s) where s = 12 for monthly seasonality
model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)

# -----------------------
# 3. Forecast Next 12 Months
# -----------------------
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# -----------------------
# 4. Visualization
# -----------------------
plt.figure(figsize=(12,6))
plt.plot(ts, label="Observed")
plt.plot(forecast_mean, label="Forecast", color="red")
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)

plt.title("SARIMA Forecast - Airline Passengers")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

Question = 9 >>> Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)

```python
# ==============================================
# Local Outlier Factor (LOF) Anomaly Detection
# ==============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import LocalOutlierFactor

# -----------------------
# 1. Generate Synthetic Data
# -----------------------
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.6, random_state=42)

# Add some random outliers
rng = np.random.RandomState(42)
outliers = rng.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, outliers])

# -----------------------
# 2. Apply LOF
# -----------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)   # -1 = anomaly, 1 = normal

# -----------------------
# 3. Visualization
# -----------------------
plt.figure(figsize=(8,6))

# Normal points
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], 
            c='blue', label='Normal', alpha=0.6)

# Anomalies
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], 
            c='red', label='Anomaly', alpha=0.8, marker='x')

plt.title("Local Outlier Factor (LOF) - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Ans = 1. Anomaly Detection in Streaming Data

Since data comes in real-time, anomalies = unexpected spikes/drops in consumption.

Options:

Isolation Forest

Works well on high-dimensional data (usage + weather + region).

Detects global anomalies.

Example: Sudden spike in usage not explainable by temperature/region.

Local Outlier Factor (LOF)

Detects local anomalies (compared to neighbors).

Useful if some regions normally have high usage but one substation suddenly deviates.

DBSCAN

Good for clustering-based anomaly detection.

Can mark small groups of unusual usage as anomalies.

Less suited for real-time streaming since it‚Äôs batch-oriented.

# Choice in practice:

Use Isolation Forest (fast, scalable, online-friendly).

Possibly combine with LOF for local checks in specific regions.

Integrate with streaming frameworks (e.g., Apache Kafka + Spark Streaming).

Options:

ARIMA ‚Üí captures trend but no seasonality/external factors.

SARIMA ‚Üí captures trend + seasonality (daily/weekly cycles).

SARIMAX ‚Üí SARIMA + exogenous features (weather, holidays, region).

# Choice in practice:

Use SARIMAX, because weather and regional effects strongly influence energy usage.
